```python
from sklearn.datasets import load_boston
boston  = load_boston()
X_data = boston.data 
print X_data
```

```python
Y_target = boston.target
print Y_target
```

```python
#將數據分成Train Set & Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_target,test_size=0.2, random_state=3)
```

```python
import time
from sklearn.linear_model import LinearRegression
model = LinearRegression()
start = time.clock()
model.fit(X_train,Y_train)

train_score = model.score(X_train,Y_train) #統計模型針對訓練樣本的準確性得分(對訓練樣本擬合的好壞程度)
cv_score = model.score(X_test,Y_test) #統計模型針對測試樣本的準確性得分(對測試樣本擬合的好壞程度)

print ('elaspe:{0:.6f}; train_score:{1: 0.6f}; cv_score: {2: .6f}'.format(time.clock()-start,train_score,cv_score))
```
